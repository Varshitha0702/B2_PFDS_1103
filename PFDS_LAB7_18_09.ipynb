```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/daily_temperature.csv")

# Ensure 'Day' is parsed as datetime (important for plotting)
df['Day'] = pd.to_datetime(df['Day'])

# Plot line chart
plt.figure(figsize=(10,5))
plt.plot(df['Day'], df['Temperature'], marker='o', linestyle='-', color='blue')

# Add labels, title, and grid
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Variation - January 2025")
plt.grid(True)
plt.xticks(rotation=45)

# Show chart
plt.tight_layout()
plt.show()

# Interpretation
hottest_day = df.loc[df['Temperature'].idxmax()]
coldest_day = df.loc[df['Temperature'].idxmin()]

print("Hottest Day:", hottest_day['Day'].date(), "with", hottest_day['Temperature'], "°C")
print("Coldest Day:", coldest_day['Day'].date(), "with", coldest_day['Temperature'], "°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/sales_data (1).csv")

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df['Quarter'], df['Sales'], marker='o', linestyle='-', color='blue')

# Highlight peak sales
peak_idx = df['Sales'].idxmax()
plt.scatter(df['Quarter'][peak_idx], df['Sales'][peak_idx],
            color='red', s=100, label='Peak Sales')

# Add labels, title, grid
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("Quarterly Sales Performance - 2025")
plt.grid(True)
plt.legend()

plt.show()

# Discussion
peak_quarter = df.loc[peak_idx, 'Quarter']
peak_sales = df.loc[peak_idx, 'Sales']

print(f"Peak sales occurred in {peak_quarter} with {peak_sales} units.")
```

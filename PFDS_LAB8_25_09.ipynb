# Compare the population of top 5 countries

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/movies_genre.csv")   # CSV should have columns: Genre, Movie_Count

# Sort by Movie Count (optional, for better visualization)
data_sorted = data.sort_values(by="Movie_Count", ascending=False)

# Plot bar chart
plt.figure(figsize=(10,6))
bars = plt.bar(data_sorted["Genre"], data_sorted["Movie_Count"], color="lightcoral", edgecolor="black")

# Add labels and title
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Number of Movies Produced by Genre")
plt.xticks(rotation=45, ha="right")

# Annotate values on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# Visualize the count of movies produced in each genre

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Load dataset
data = pd.read_csv("/content/population.csv")  # Ensure your CSV has columns: Country, Population

# Sort and take top 5 countries
top5 = data.sort_values(by="Population", ascending=False).head(5)

# Plot bar chart
plt.figure(figsize=(8,6))
bars = plt.bar(top5["Country"], top5["Population"], color="skyblue", edgecolor="black")

# Format y-axis (millions/billions)
def billions(x, pos):
    if x >= 1e9:
        return f'{x*1e-9:.1f}B'
    elif x >= 1e6:
        return f'{x*1e-6:.1f}M'
    else:
        return f'{x:.0f}'

plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(billions))

# Labels and title
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 5 Most Populated Countries")
plt.xticks(rotation=45)

# Annotate values on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             billions(height, None), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
